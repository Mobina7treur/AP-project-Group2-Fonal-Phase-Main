# "Annotated Heatmaps":
## Application: An annotated heatmap provides a clear visual representation of data, where annotations (e.g., average Final_Price by Category and Payment_Method) are displayed inside each cell. 
### Description: A pivot table has been used here to calculate the mean final price and displays the values within each cell

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a pivot table for the heatmap (e.g., average Final_Price by Category and Payment_Method)
pivot_table = ecommerce.pivot_table(
    values='Final_Price(Rs.)', 
    index='Category', 
    columns='Payment_Method', 
    aggfunc=np.mean
)

# Plot an annotated heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    pivot_table, 
    annot=True,        # Annotate with numeric values
    fmt=".2f",         # Format for annotations (2 decimal places)
    cmap='coolwarm',   # Color palette
    linewidths=0.5,    # Line width between cells
    linecolor='black', # Line color
    cbar_kws={'label': 'Average Final Price (Rs.)'}
)
plt.title('Annotated Heatmap: Average Final Price by Category and Payment Method', fontsize=16)
plt.xlabel('Payment Method', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.tight_layout()
plt.show()
```

# "Multiple bivariate KDE (Kernel Density Estimate)"
## Application: By sketching multiple bivariate KDE (Kernel Density Estimate) plots, we can compare the relationships between several pairs of variables in the dataset.
### Description: Bivariate KDE plots here have three variable pairs, visualizing their density distributions

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define pairs of variables for bivariate KDE plots
pairs = [
    ('Price (Rs.)', 'Final_Price(Rs.)'),
    ('Price (Rs.)', 'Discount (%)'),
    ('Final_Price(Rs.)', 'Discount (%)')
]

# Create subplots
fig, axes = plt.subplots(1, len(pairs), figsize=(18, 6), sharex=False, sharey=False)

for i, (x_var, y_var) in enumerate(pairs):
    sns.kdeplot(
        x=ecommerce[x_var], 
        y=ecommerce[y_var], 
        ax=axes[i], 
        cmap='viridis', 
        fill=True, 
        thresh=0, 
        levels=100
    )
    axes[i].set_title(f'{x_var} vs {y_var}', fontsize=14)
    axes[i].set_xlabel(x_var, fontsize=12)
    axes[i].set_ylabel(y_var, fontsize=12)

plt.tight_layout()
plt.show()
```

## 2. <u>Identifying unusual purchasing patterns or suspicious transactions that may indicate fraud:</u>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Selecting relevant features for clustering (Purchase Amount & Discount)
clustering_data = ecommerce[["Final_Price(Rs.)", "Discount (%)"]]

# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Applying K-Means Clustering
num_clusters = 3  # Assuming 3 main clusters (Normal, Discount-Seeking, Fraudulent)
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
ecommerce["Cluster"] = kmeans.fit_predict(scaled_data)

# Plot Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=ecommerce,
    x="Discount (%)",
    y="Final_Price(Rs.)",
    hue=ecommerce["Cluster"].astype(str),
    palette="viridis",
    s=50
)

plt.title("Cluster Analysis of Purchase Amounts and Discounts")
plt.xlabel("Discount (%)")
plt.ylabel("Final Price (Rs.)")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

# <span style="color:brown;"> How the Cluster Analysis Shows Unusual Purchasing Patterns:</span>

# ➤In the **cluster analysis plot**, we grouped purchases based on:
### 1. **Discount Percentage** (X-axis) → How much discount was applied.
### 2. **Final Price (Rs.)** (Y-axis) → How much the customer paid after the discount.

# ➤ **How We Detect Unusual Purchasing Patterns**
### - We used **K-Means clustering** to group similar purchases into **three categories (clusters)**.
### - Each cluster is shown with **a different color** in the scatter plot.
### - **Unusual purchases** are found by looking at the cluster that **does not follow the general trend**.

---

# ➤ **Identifying Suspicious Transactions**
## 1. **Regular Purchases (Normal Cluster)**
####   - Most purchases fall in **a compact group** with reasonable discounts and final prices.
#### - These represent **typical shopping behavior**.

## 2. **Heavy Discount Seekers (Second Cluster)**
#### - These transactions have **very high discounts** (e.g., 80%-90%) but **low final prices**.
#### - Could be **aggressive sale shoppers** or **potential misuse of discount codes**.

## 3. **Suspicious Transactions (Third Cluster - Unusual)**
### - Purchases where:
#### - **Extremely high discounts** were applied, but the **final price is still high**.
#### - This could indicate **fraudulent discount manipulation** (e.g., unauthorized coupon stacking).
#### - If one user repeatedly gets **huge discounts on expensive products**, it might be **fraud**.

---

### - **Unusual transactions appear as a separate group** that is far from the regular clusters.
### - If a purchase **gets an extreme discount but still has a high final price**, it **might be fraud**.
### - This visualization helps **flag these suspicious purchases for further investigation**.

# <span style="color:brown;">Analysis of the "wraping up diagram":</span>
## ✅As mentioned earlier, we sketched this diagram to observe the percentage distribution of the categories among all the sold products. Now, we will analyze it more accurately <u>to suggest some customer classification methods.</u>
##  ➤1.Prioritizing Larger Categories:
### The highest percentages (e.g., Home & Kitchen, Books) could suggest that more attention can be focoused on customers who tend to purchase those products, since they represent a larger part of the total sale.
##  ➤2. Purchase Frequency Segmentation:
### Identify customers who frequently purchase from the larger categories, like Home & Kitchen and Books. These customers can be classified into segments such as "Frequent Buyers" and "Occasional Shoppers" based on the number of times they purchase from these categories. This allows for targeting more loyal customers with exclusive deals or rewards.

##  ➤3. Basket Composition Analysis(Cross-Category Shopper) : 
### Rather than focusing only on individual category preferences, analyze which products tend to be purchased together. For example, if a customer who buys from the "Electronics" category often purchases from "Sports," they can be classified as a "Cross-Category Shopper." This segmentation enables the creation of bundled offers, promoting products from both categories together.

##  ➤4. Seasonal Behavior Segmentation:
### Depending on the product categories, seasonal trends may emerge (e.g., Home & Kitchen or Books might see higher sales during certain holidays). Customers can be classified into "Seasonal Shoppers" or "Year-Round Shoppers." This approach helps optimize marketing efforts to target customers who are more active during specific seasons, resulting in more focused and effective campaigns.
